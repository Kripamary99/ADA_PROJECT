Here the field 'stress_trig' has multiple combinations of the options specified in the form. Simply label encoding everything will give diiferent values for the same(for example: some entry will have 'Academic, Family' and some will have 'Academic, Social' both has 2 reasons for stress triggering but the label encoder will encode both as different). Hence we calculate the commas in each entry and add 1 to it (for example: if one entry has only 'Academic' the code will calculate the commas in it, since there is no comma it will be taken as 0 and 1 will be added to it, if one entry has 'Academic, Family' there is 1 comma and it will be taken and after adding 1 the value of that entry will be 2, since there are 2 reasons for stress). Hence that we would be able to calculate the amount of factors.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
import math
lr=LinearRegression()
y=df[['stress_level']]
df.drop('stress_level',axis=1,inplace=True)
x=df
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=2)
lr.fit(xtrain,ytrain)
```

MULTI-LINEAR REGRESSION

```python
ypred=lr.predict(xtest)
mse_lr=mean_squared_error(ytest,ypred)
print("MSE: ",mse_lr)
mae_lr=mean_absolute_error(ytest,ypred)
print("MAE: ",mae_lr)
rmse_lr = math.sqrt(mse_lr)
print("RMSE: ",rmse_lr)
r2 = r2_score(ytest, ypred)
print("R-squared:", r2)
n = xtest.shape[0]
p = xtest.shape[1]
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2)
print(lr.coef_)
print(lr.intercept_)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain, ytrain)
y_pred_rf = rf.predict(xtest)
mse_rf = mean_squared_error(ytest, y_pred_rf)
print("Mean Squared Error (MSE):", mse_rf)
mae_rf = mean_absolute_error(ytest, y_pred_rf)
print("Mean Absolute Error (MAE):", mae_rf)
rmse_rf = math.sqrt(mse_rf)
print("Root Mean Squared Error (RMSE):", rmse_rf)
r2_rf = r2_score(ytest, ypred)
print("R-squared:", r2_rf)
n = xtest.shape[0]
p = xtest.shape[1]
adjusted_r2_rf = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2_rf)
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeRegressor
dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(xtrain, ytrain)
y_pred_dt = dt_regressor.predict(xtest)
mse_dt = mean_squared_error(ytest, y_pred_dt)
print("Mean Squared Error (MSE):", mse_dt)
mae_dt = mean_absolute_error(ytest, y_pred_dt)
print("Mean Absolute Error (MAE):", mae_dt)
rmse_dt = math.sqrt(mse_dt)
print("Root Mean Squared Error (RMSE):", rmse_dt)
r2_dt = r2_score(ytest, y_pred_dt)
print("R-squared:", r2_dt)
n = xtest.shape[0]
p = xtest.shape[1]
adjusted_r2_dt = 1 - (1 - r2_dt) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2_dt)
```

RIDGE RESSION

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(xtrain, ytrain)
y_pred_ridge = ridge.predict(xtest)
mse_ridge = mean_squared_error(ytest, y_pred_ridge)
print("Mean Squared Error (MSE):", mse_ridge)
mae_ridge = mean_absolute_error(ytest, y_pred_ridge)
print("Mean Absolute Error (MAE):", mae_ridge)
rmse_ridge = math.sqrt(mse_ridge)
print("Root Mean Squared Error (RMSE):", rmse_ridge)
r2_ridge = r2_score(ytest, y_pred_ridge)
print("R-squared:", r2_ridge)
n = xtest.shape[0]
p = xtest.shape[1]
adjusted_r2_ridge = 1 - (1 - r2_ridge) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2_ridge)
```

LASSO REGRESSION

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(xtrain, ytrain)
y_pred_lasso = lasso.predict(xtest)
mse_lasso = mean_squared_error(ytest, y_pred_lasso)
print("Mean Squared Error (MSE):", mse_lasso)
mae_lasso = mean_absolute_error(ytest, y_pred_lasso)
print("Mean Absolute Error (MAE):", mae_lasso)
rmse_lasso = math.sqrt(mse_lasso)
print("Root Mean Squared Error (RMSE):", rmse_lasso)
r2_lasso = r2_score(ytest, y_pred_lasso)
print("R-squared:", r2_lasso)
n = xtest.shape[0]
p = xtest.shape[1]
adjusted_r2_lasso = 1 - (1 - r2_lasso) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2_lasso)
```

```python
metrics = {
    'Model': ['Linear Regression', 'Random Forest', 'Decision Tree', 'Polynomial Regression', 'Lasso Regression'],
    'MSE': [mse_lr, mse_rf, mse_dt, mse_ridge, mse_lasso],
    'MAE': [mae_lr, mae_rf, mae_dt, mae_ridge, mae_lasso],
    'RMSE': [rmse_lr, rmse_rf, rmse_dt, rmse_ridge, rmse_lasso],
    'R-squared': [r2, r2_rf, r2_dt, r2_ridge, r2_lasso],
    'Adjusted R-squared': [adjusted_r2, adjusted_r2_rf, adjusted_r2_dt, adjusted_r2_ridge, adjusted_r2_lasso]
}
metrics_df = pd.DataFrame(metrics)
metrics_df.set_index('Model', inplace=True)
```
